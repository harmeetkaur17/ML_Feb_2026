
Accessing and updating list:

```python
print("ls1[0]=",ls1[0]) #access an item at index 0
print("ls1[-2]=",ls1[-2]) #access an item at index -2
print("ls1[1:3]=",ls1[1:3])  ## access a range of items at indices from 0 to 3 
ls1[1]="Hello" ##update one item of the list
print("ls1=", ls1) 
ls1[2:4]=[4,20] ## update multiple items of the list
print("ls1=", ls1)
```

Delete list items

```python
ls=[2,19,3, "hello", 3, 9.2]
print("ls=", ls)
del ls[1]  # delete the list item (19) at index 1 
print("ls=", ls)
ls.remove(3) # will remove the first matching number (3) from list
print("ls=", ls)
```

List operators:

```python
ls1=[2,7,8]
ls2=["hi","hello", "welcome"]
print("ls1+ls2:", ls1+ls2) # concatenation
print("ls1*3:", ls1*3) # repetition
print("7 in ls1:", 7 in ls1) # membership
print('hi" not in ls2:', "hi" not in ls2) # membership
```


Built-in List methods:

```python
ls=[2,19,3, "hello", 3, 9.2]
print("ls=", ls)
print("len(ls)=", len(ls)) # length of list
ls.append(20)  #Adds an item to the list
print("ls=",ls)
ls2=["hi", "welcome"]
ls.extend(ls2) # Adds squence to the list
print("ls=",ls)
 
ls3=[2,8,2,11,1,7]
print("max(ls3)=", max(ls3)) # maximum value of the list
print("min(ls3)=", min(ls3)) # minimum value of the list
ls3.sort() # sort list in ascending order
print("ls3.sort():", ls3) 
ls3.insert(2,15) # insert value at a specific index
print("ls3.insert(15,2):",ls3)
print("ls3.index(15)=", ls3.index(15)) # get index
```

Updating a dictionary

```python

dict1= {"h": "hello","w": "welcome","r":"Robogarden"}
print("dict1=", dict1)
	
dict1["j"]="John" # adding a new entry to the dictionary
print("dict1=", dict1)
 
dict1['h']="hi" # modify an existing entry 
print("dict1=", dict1)
 
del dict1['w']  # delete an existing entry
print("dict1=", dict1)
 
dict1.clear()  # remove all entries in the dictionary
print("dict1=", dict1)
 
del dict1  # delete the dictionary itself
#print("dict1=", dict1)
```

```python
dict1={1:"Hi", 2:"welcome",3:"RoboGarden"}
print("dict1=", dict1)
print("len(dict1):", len(dict1))
print("dict1.items():", dict1.items())
print("dict1.values():", dict1.values())
print("dict1.keys():", dict1.keys())
print("output sting=", "The dict1 is "+str(dict1))
print("dict1.get(key, default=False): ", dict1.get(7, False))
print("dict1.get(key, default=False): ", dict1.get(2, False))

print("dict1.get(key, default=False): ", dict1.get(3, False))
 
 
dict1.setdefault(4,"everyone") #returns the value if exist otherwise insert the value
print("After setting default: dict1=", dict1)
 
dict2=dict1.copy()
print("dict2=", dict2)
 
dict3={8:"yes", 9:"No"}
dict1.update(dict3)
print("dict1=", dict1)
```
